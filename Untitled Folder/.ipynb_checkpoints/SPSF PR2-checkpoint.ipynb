```python
import numpy as np
import matplotlib.pyplot as plt


def rungeKutta(F, t, x, y, h):
    K0 = h * F(t, x, y)
    K1 = h * F(t + h/2.0, x + K0/2.0, y)
    K2 = h * F(t + h/2.0, x + K1/2.0, y)
    K3 = h * F(t + h, x + K2, y)
    return(K0 + 2.0 * K1 + 2.0 * K2 + K3)/6.0


def vx_terkopel(t, vx, vy, q = 1, Bz = 1, m = 1):
    return -q * Bz/m * vy

def vy_terkopel(t, vy, vx, q = 1, Bz = 1, m = 1):
    return q * Bz/m * vx
```

```python
#Variable Assignment
vx0 = 0
vy0 = 20 
t0 = 0
t_stop = 30
h = 0.005

#Array Declaration
vxs=[vx0]
vys=[vy0]
ts=[t0]
```

```python
while ts[-1] < t_stop:
    h = min(h, t_stop - ts[-1])
    vx = vxs[-1] + rungeKutta(vx_terkopel,ts[-1],vxs[-1],vys[-1],h)
    vy = vys[-1] + rungeKutta(vy_terkopel,ts[-1],vys[-1],vxs[-1],h)
    t = ts[-1] + h
    
    ts.append(t)
    vxs.append(vx)
    vys.append(vy)

plt.figure()
plt.grid()
plt.plot(ts, vys, 'red', label = '$v_x$')
plt.plot(ts, vxs, 'black', label = '$v_y$')
plt.title('Rotasi partikel dengan Runge-Kutta')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan')
plt.legend(loc = 'upper left')

plt.show()
```

```python
#Solusi Analitik
t = np.arange(0, 30, 0.05)

v_x = -20 * np.sin(t)
v_y = 20 * np.cos(t)

plt.figure()
plt.grid()
plt.plot(t, v_x, 'red', label = '$v_x$')
plt.plot(t, v_y, 'black', label = '$v_y$')
plt.title('Rotasi Partikel Secara Teori')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan')
plt.legend(loc = 'upper left')

plt.show()
```
