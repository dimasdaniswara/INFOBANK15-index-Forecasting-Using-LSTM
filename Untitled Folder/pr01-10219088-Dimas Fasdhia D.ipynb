```python
from numpy import *
from matplotlib.pyplot import *
```

```python
#x’ = F_1, y’ = F_2
def F(t,x,y):
    return (-x + x*y, y - x*y)


#Forward Euler’s Method for 2 dimensions
def euler2D(t0,tn,n,x0,y0):
    h = abs(tn-t0)/n
    t = linspace(t0,tn,n+1)
    x = zeros(n+1)
    y = zeros(n+1)
    x[0] = x0
    y[0] = y0
    for k in range(0,n):
        (dx,dy) = F(t[k],x[k],y[k])
        x[k+1] = x[k] + h*dx
        y[k+1] = y[k] + h*dy
    return (x,y)

#Runge-Kutta "Classic" Order 4 method for 2 dimensions
def RK42D(t0,tn,n,x0,y0):
    h = abs(tn-t0)/n
    t = linspace(t0,tn,n+1)
    x = zeros(n+1)
    y = zeros(n+1)
    x[0] = x0
    y[0] = y0
    for k in range(0,n):
        (dx1,dy1) = F(t[k],x[k],y[k])
        (dx2,dy2) = F(t[k]+h/2,x[k]+dx1*h/2,y[k]+dy1*h/2)
        (dx3,dy3) = F(t[k]+h/2,x[k]+dx2*h/2,y[k]+dy2*h/2)
        (dx4,dy4) = F(t[k]+h, x[k]+dx3*h, y[k]+dy3*h)
        x[k+1] = x[k] + h*(dx1 +2*dx2 +2*dx3 +dx4)/6
        y[k+1] = y[k] + h*(dy1 +2*dy2 +2*dy3 +dy4)/6
    return (x,y)

#Input values
t0 = 0
tn = 20
n = 1000
x0 = 0.3
y0 = 2
#Plotting Commands for Euler approximation
figure(1) #x vs y plot
(xpt,ypt) = euler2D(t0,tn,n,x0,y0)
plot(xpt,ypt,color='blue',label='Euler')
title('Predator vs Prey')
xlabel('predator population')
ylabel('prey population')
axis([-1,5,-1,5])
```

```python
figure(2) #x vs t and y vs t subplots
t = linspace(t0,tn,n+1)
subplot(211)
title('Predator/Prey populations over time')
ylabel('predator')
plot(t,xpt,color='blue')X
subplot(212)
xlabel('time')
ylabel('prey')
plot(t,ypt,color='blue')
#Plotting Commands for Runge-Kutta 4 approximation
figure(1) #x vs y plot
(xpt,ypt) = RK42D(t0,tn,n,x0,y0)
plot(xpt,ypt,color='red',label='RK4')
legend(loc='best')
figure(2) #x vs t and y vs t subplots
subplot(211)
plot(t,xpt,color='red')
subplot(212)
plot(t,ypt,color='red')
```
