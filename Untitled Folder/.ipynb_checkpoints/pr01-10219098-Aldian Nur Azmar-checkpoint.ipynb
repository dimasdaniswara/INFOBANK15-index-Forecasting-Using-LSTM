```python
# Mendefinisikan library
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(t, x, y):
    return (-x + (x * y), y - (x * y))
```

```python
# Mendefinisikan euler method
def eulerMethod(t_0, t_n, num_steps, x_0, y_0):
    h = abs(t_n - t_0)/num_steps
    t = np.linspace(t_0, t_n, num_steps + 1)
    x = np.zeros(num_steps + 1)
    y = np.zeros(num_steps + 1)
    x[0] = x_0
    y[0] = y_0
    for k in range(0, num_steps):
        (dx, dy) = F(t[k], x[k], y[k])
        x[k + 1] = x[k] + h * dx
        y[k + 1] = y[k] + h * dy
    return (x, y)
```

```python
# Mendefinisikan Runge Kutta method
def rungeKutta(t_0, t_n, num_steps, x_0, y_0):
    h = abs(t_n - t_0)/num_steps
    t = np.linspace(t_0, t_n , num_steps + 1)
    x = np.zeros(num_steps + 1)
    y = np.zeros(num_steps + 1)
    x[0] = x_0
    y[0] = y_0
    for k in range(0, num_steps):
        (x1, y1) = F(t[k], x[k], y[k])
        (x2, y2) = F(t[k] + h/2, x[k] + x1 * h/2, y[k] + y1 * h/2)
        (x3, y3) = F(t[k] + h/2, x[k] + x2 * h/2, y[k] + y2 * h/2)
        (x4, y4) = F(t[k] + h, x[k] + x3 * h, y[k] + y3*h)
        x[k + 1] = x[k] + h * (x1 + 2 * x2 + 2 * x3 + x4)/6
        y[k + 1] = y[k] + h * (y1 + 2 * y2 + 2 * y3 + y4)/6
    return (x,y)
```

```python
# Input nilai parameter yang dibutuhkan
t_0 = 0
t_n = 100
x_0 = 0.3
y_0 = 2
num_steps = 3000
```

```python
# Plotting grafik yang diperoleh berdasarkan
# persamaan euler untuk visualisasi predator - prey
plt.figure(1)
(x_pt,y_pt) = eulerMethod(t_0, t_n, num_steps, x_0, y_0)
plt.plot(x_pt,y_pt,color='blue',label='Euler')
plt.title('Grafik Populasi Predator Terhadap Mangsa - Persamaan Euler')
plt.xlabel('Populasi Predator')
plt.ylabel('Populasi Mangsa')
plt.axis([-1, 5, -1, 5])
plt.savefig('Figure (1) Grafik Predator - Prey berdasarkan Persamaan Euler.png')
```

```python
# Visualisasi grafik hubungan Predator - Prey yang ditinjau setiap waktu
# berdasarkan Persamaan Euler
plt.figure(2)
t = np.linspace(t_0, t_n, num_steps + 1)
plt.subplot(2, 1, 1)
plt.title('Populasi Predator dan Mangsa Setiap Waktu')
plt.ylabel('Jumlah Predator')
plt.plot(t,x_pt,color='red')

plt.subplot(2, 1, 2)
plt.xlabel('Waktu (s)')
plt.ylabel('Jumlah Mangsa')
plt.plot(t,y_pt,color='blue')

plt.savefig('Figure (2) Grafik Populasi Predator - Prey yang ditinjau setiap waktu (Persamaan Euler).png')
```

```python
# Plotting grafik yang diperoleh berdasarkan
# persamaan Runge - Kutta untuk visualisasi predator - prey
plt.figure(1)
(xpt, ypt) = rungeKutta(t_0, t_n, num_steps, x_0, y_0)
plt.plot(xpt,ypt,color='red',label='Runge-Kutta')
plt.title('Grafik Populasi Predator Terhadap Mangsa - Persamaan Runge-Kutta')
plt.ylabel('Populasi Mangsa')
plt.xlabel('Populasi Predator')
plt.legend(loc='best')
plt.savefig('Figure (3) Grafik Predator - Prey berdasarkan Persamaan Runge-Kutta.png')
```

```python
# Mendefinisikan fungsi Lotka–Volterra Model
def LotkaVolterraModel(x,params):
    alpha = params["alpha"]
    beta = params["beta"]
    gamma = params["gamma"]
    delta = params["delta"]
    
    xdot = np.array([alpha*x[0] - beta*x[0]*x[1], delta*x[0]*x[1] - gamma*x[1]])
    return xdot
```

```python
def RungeKutta(f, x0, t0, tf, dt):
    
    t = np.arange(t0, tf, dt)
    nt = t.size
    
    nx = x0.size
    x = np.zeros((nx,nt))
    
    x[:,0] = x0
    
    for k in range(nt - 1):
        k1 = dt*f(t[k], x[:,k])
        k2 = dt*f(t[k] + dt/2, x[:,k] + k1/2)
        k3 = dt*f(t[k] + dt/2, x[:,k] + k2/2)
        k4 = dt*f(t[k] + dt, x[:,k] + k3)
        
        dx = (k1 + 2*k2 + 2*k3 + k4)/6
        x[:,k+1] = x[:,k] + dx
    
    return x,t
```

```python
# Memanggil fungsi Lotka-Volterra Model untuk
# input pada persamaan Runge-Kutta
f = lambda t, x :LotkaVolterraModel(x,params)
x0 = np.array([20, 5])
params = {"alpha": 1.1, "beta": 0.4, "gamma": 0.4, "delta": 0.1}
```

```python
# Memanggil fungsi Runge-Kutta
t0 = 0
tf = 100.0
dt = 0.005
x, t = RungeKutta(f, x0, t0, tf, dt)
```

```python
# Visualisasi grafik hubungan Predator - Prey yang ditinjau setiap waktu
# berdasarkan Persamaan Runge-Kutta
plt.subplot(1, 2, 1)
plt.plot(t, x[0,:], "r", label = "Preys")
plt.plot(t, x[1,:], "y", label = "Predators")
plt.xlabel("Time (t)")
plt.title('Populasi Predator dan Mangsa Setiap Waktu')
plt.grid()
plt.legend()
plt.savefig('Figure (4) Grafik Populasi Predator - Prey yang ditinjau setiap waktu (Runge-Kutta Method)')
plt.show()
```
